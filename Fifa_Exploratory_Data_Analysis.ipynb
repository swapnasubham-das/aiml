## Data Visualization

```python
count_preferred_foot_df = fifa_data['Preferred Foot'].value_counts().reset_index()
# Rename the columns
count_preferred_foot_df.columns = ['Preferred Foot', 'Count']
count_preferred_foot_df
```

#### Number of Players by Preferred Foot

```python
import plotly.express as px
# Create the bar chart using Plotly Express
fig = px.bar(fifa_data['Preferred Foot'].value_counts().reset_index(),x='Preferred Foot',y='count',
    labels={'index': 'Preferred Foot', 'Preferred Foot': 'Number of Players'},title='Number of Players by Preferred Foot'
)

# Update layout for better appearance
fig.update_layout(
    xaxis_title='Preferred Foot',
    yaxis_title='Number of Players',
    template='plotly_white'
)

# Add value labels on top of each bar
fig.update_traces(texttemplate='%{y}', textposition='outside')

# Show the plot
fig.show()
```

#### Percentage wise Preferred Foot

```python
percentage_preferred_foot_df = pd.DataFrame(fifa_data['Preferred Foot'].value_counts() / len(fifa_data)).reset_index()
#Rename the columns
percentage_preferred_foot_df.columns = ['Preferred Foot', 'Percentage']
percentage_preferred_foot_df
```

```python
fig = px.pie(percentage_preferred_foot_df,values='Percentage',names='Preferred Foot',
             title='Distribution of Preferred Foot amongs Players')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

#### Players Age distribution

```python
import plotly.express as px
fig = px.histogram(fifa_data, x="Age",title='Age distribution of Players')
fig.show()
```

```python
import plotly.figure_factory as ff

hist_data = [fifa_data['Age']]
group_labels = ['Age distribution of Players'] # name of the dataset
fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```

#### International Reputation of Players

```python
percentage_international_reputation_df = pd.DataFrame(fifa_data['International Reputation'].value_counts().reset_index())
#Rename the columns
percentage_international_reputation_df.columns = ['International Reputation', 'Total']
percentage_international_reputation_df
```

```python
fig = px.pie(percentage_international_reputation_df,values='Total',names='International Reputation',
             title='Distribution of International Reputation of Players')
fig.show()
```

#### International Reputation by Preferred Foot

```python
preferred_foot_international_reputation = fifa_data.groupby('Preferred Foot')['International Reputation'].value_counts().reset_index()
preferred_foot_international_reputation

# Create a bar plot
fig = px.bar(preferred_foot_international_reputation, x='International Reputation', y='count', color='Preferred Foot',
             barmode='group', title='International Reputation by Preferred Foot',
             labels={'count': 'Count', 'International Reputation': 'International Reputation'})

# Show the plot
fig.show()
```

```python
position_mapping = {
    'ST': 'Striker',
    'GK': 'Goalkeeper',
    'CB': 'Center Back',
    'CM': 'Central Midfielder',
    'LB': 'Left Back',
    'RB': 'Right Back',
    'RM': 'Right Midfielder',
    'LM': 'Left Midfielder',
    'CAM': 'Central Attacking Midfielder',
    'CDM': 'Central Defensive Midfielder',
    'RCB': 'Right Center Back',
    'LCB': 'Left Center Back',
    'LCM': 'Left Central Midfielder',
    'RCM': 'Right Central Midfielder',
    'LW': 'Left Winger',
    'RW': 'Right Winger',
    'RDM': 'Right Defensive Midfielder',
    'LDM': 'Left Defensive Midfielder',
    'LS': 'Left Striker',
    'RS': 'Right Striker',
    'RWB': 'Right Wing Back',
    'LWB': 'Left Wing Back',
    'CF': 'Center Forward',
    'LAM': 'Left Attacking Midfielder',
    'RAM': 'Right Attacking Midfielder',
    'RF': 'Right Forward',
    'LF': 'Left Forward'
}
```

```python
fifa_data['Position'] = fifa_data['Position'].replace(position_mapping)
```

```python
fifa_data['Position'].value_counts()
```

```python
import plotly.express as px
# Create the bar chart using Plotly Express
fig = px.bar(fifa_data['Position'].value_counts().reset_index(),x='Position',y='count', color='Position',
    labels={'index': 'Position', 'Position': 'Position'},title='Total no of players based on Position')

# Update layout for better appearance
fig.update_layout(
    xaxis_title='Position',
    yaxis_title='Number of Players',
    template='plotly_white'
)

# Add value labels on top of each bar
fig.update_traces(texttemplate='%{y}', textposition='outside')

# Show the plot
fig.show()
```
