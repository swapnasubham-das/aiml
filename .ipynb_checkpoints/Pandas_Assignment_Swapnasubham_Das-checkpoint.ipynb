```python
missing_percentage = (netflix_df.isnull().sum() * 100 / len(netflix_df)).round(2)
missing_percentage
```

```python
df_netflix_cleaned = netflix_df.dropna(subset=['director'])
df_netflix_cleaned
```

```python
netflix_df['country'].fillna('', inplace=True)
filtered_netflix_df = netflix_df[netflix_df['country'].str.contains('Germany|West Germany', regex=True)]
filtered_netflix_df
```

```python
netflix_df[['Duration_Numeric', 'Duration_String']] = netflix_df['duration'].str.extract(r'(\d+)\s*(\w+)', expand=True)
netflix_df
```

```python
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'])
# Split into separate columns
netflix_df['day'] = netflix_df['date_added'].dt.day
netflix_df['month'] = netflix_df['date_added'].dt.month
netflix_df['year'] = netflix_df['date_added'].dt.year
netflix_df
```

```python
release_counts = netflix_df['release_year'].value_counts().sort_index()
print("Number of TV shows/Movies released in each year:")
print(release_counts)
```

```python
netflix_df = netflix_df.rename(columns={'title': 'movie_title'})
print(netflix_df.columns)
```

```python
netflix_df[['Genre1', 'Genre2', 'Genre3']] = netflix_df['listed_in'].str.split(',', expand=True, n=2)

netflix_df['Genre1'] = netflix_df['Genre1'].str.strip()
netflix_df['Genre2'] = netflix_df['Genre2'].str.strip()
netflix_df['Genre3'] = netflix_df['Genre3'].str.strip()

netflix_df.head()
```

```python
# Define a function to assign IQ levels
def assign_iq_level(score):
    if score > 80:
        return 'Intelligent'
    elif score >= 50 and score <= 80:
        return 'Mediocre'
    else:
        return 'Weak'

# Create the new 'IQ_level' column using apply()
mark_df['IQ_level'] = mark_df['Mark'].apply(assign_iq_level)
mark_df
```

```python
gender_city_count = join_df.groupby('City')['Gender'].value_counts().unstack(fill_value=0)
gender_city_count
```

```python
top_5_males = join_df[join_df['Gender'] == 'Male'].sort_values('Mark', ascending=False).head(5)
top_5_males
```

```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
        'Age': [28, 24, None, 32],
        'City': ['New York', None, 'London', 'San Francisco']}
df = pd.DataFrame(data)

# Identify missing values
print("Actual Dataframe:")
print(df)

# Identify missing values
print("\nMissing values identified:")
print(df.isnull())

# Drop rows with missing values
df_dropped_rows = df.dropna()
print("\nDataFrame after dropping rows with missing values:")
print(df_dropped_rows)

# Fill missing values with a specific value
df_filled_value = df.fillna('Unknown')
print("\nDataFrame after filling missing values with 'Unknown':")
print(df_filled_value)

# Fill missing values with the mean of the 'Age' column
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("\nDataFrame after filling missing values in 'Age' with the mean:")
print(df)

# Interpolate missing values in the 'Age' column
df['Age'] = df['Age'].interpolate()
print("\nDataFrame after interpolating missing values in 'Age':")
print(df)

# Replace None with a specific value
df_replaced = df.replace({None: 'Unknown'})
print("\nDataFrame after replacing None with 'Unknown':")
print(df_replaced)
```
